```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars_data = pd.read_csv("Toyota.csv", index_col=0, na_values=["??", "????"])
```

```python
cars_data
```

```python
cars_data.dropna(axis=0, inplace=True)
```

```python
cars_data
```

```python
plt.scatter(cars_data["Age"], cars_data["Price"], c="red")
plt.title("Scatter plot of Price vs Age of the cars")
plt.xlabel("Age(Months)")
plt.ylabel("Price")
plt.show()
```

```python
# histogram
```

```python
plt.hist(cars_data["KM"])
```

```python
plt.hist(cars_data["KM"],
        color="green",
        edgecolor="white",
        bins=5)
plt.title("Histogram of KM")
plt.xlabel("KM")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(cars_data["KM"],
        color="green",
        edgecolor="white",
        bins=5,
        orientation="horizontal")
plt.title("Histogram of KM")
plt.xlabel("KM")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar Plots
```

```python
cars_data["FuelType"].value_counts()
```

```python
counts = [968, 116, 12]
fuelType = ("Petrol", "Diesel", "CNG")
index = np.arange(len(fuelType))
```

```python
plt.bar(index, counts, color = ['red', 'blue', 'cyan'])
plt.title("Bar plot of FuelTypes")
plt.xlabel("FuelType")
plt.ylabel("Frquency(Count)")
plt.show()
```
