```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import datetime as dt
```

```python
df = pd.read_csv("/content/covid_19_india.csv", dayfirst=True)
```

```python
df.head()
```

```python
df = df[["Date","State/UnionTerritory","Cured","Deaths", "Confirmed"]]
```

```python
df.columns = ["date", "state","cured", "deaths", "confirmed" ]
```

```python
df.head()
```

```python
df.tail()
```

```python
today = df[df.date == "2021-08-11"]
```

```python
today.head()
```

```python
#sort data w.r.t number of confirmed cases
max_confirmed_cases = today.sort_values(by="confirmed", ascending = False)
```

```python
max_confirmed_cases
```

```python
#Getting states with maximum number of confirmed cases
top_state_confirmed = max_confirmed_cases[0:5]
```

```python
top_state_confirmed
```

```python
#Making bar plot for state with top confirmed cases
```

```python
sns.set(rc = {"figure.figsize":(7,4)})
```

```python
sns.barplot(x = "state", y = "confirmed", data =top_state_confirmed, hue = "state" )
```

```python
#sort data w.r.t number of confirmed cases
max_deaths_cases = today.sort_values(by="deaths", ascending = False)
```

```python
max_deaths_cases
```

```python
top_state_deaths = max_deaths_cases[0:5]
top_state_deaths
sns.set(rc = {"figure.figsize":(7,4)})
sns.barplot(x = "state", y = "deaths", data =max_deaths_cases, hue = "state")
```
